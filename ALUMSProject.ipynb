```python
import requests
from bs4 import BeautifulSoup
```

```python
site = "https://en.wikipedia.org"
page = requests.get("https://en.wikipedia.org/wiki/List_of_University_of_Illinois_at_Urbana%E2%80%93Champaign_people")
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
li = soup.find_all('li')[47:743]
```

```python
alum_links = []
for x in range(len(li)):
    if li[x].get_text():
        alum_links.append(li[x].find('a')['href'])
```

```python
alum_full_links = []
for index in range(len(alum_links)):
    alum_full_links.append(site + alum_links[index])
```

```python
names = []
fields = []
birthplaces = []
print("Name" + "+" + "Link" + "+" + "Fields" + "+" + "City" + "+" + "State" + "+" + "Country")
for y in range(len(alum_full_links)):
    page2 = requests.get(alum_full_links[y])
    soup2 = BeautifulSoup(page2.content, 'html.parser')
    div = soup2.find_all('div', class_ = 'birthplace')
    divfn = soup2.find_all('div', class_ = 'fn')
    tdcategory = soup2.find_all('td', class_ = 'category')[1:2]
    
    for z in range(len(div)):
        if div[z].get_text():
            birthplaces.append(div[z].get_text())
            break
    if len(birthplaces) < y+1:
        birthplaces.append("N/A")
        
    for z in range(len(divfn)):
        if divfn[z].get_text():
            names.append(divfn[z].get_text())
            break
    if len(names) < y+1:
        names.append("N/A")
        
    for z in range(len(tdcategory)):
        if tdcategory[z].get_text():
            fields.append(tdcategory[z].get_text())
            break
    if len(fields) < y+1:
        fields.append("N/A")
        
    print(names[y] + "+" + alum_full_links[y] + "+" + fields[y] + "+" + birthplaces[y])
```
